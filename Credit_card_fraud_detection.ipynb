importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a Pandas dataframe
credit_card_data = pd.read_csv('/creditcard.csv', on_bad_lines='skip')
```

```python
#print first 5 rows of the dataset
credit_card_data.head()
```

#dataset provider cannot provide the sensetive information about the credit card details that's why he is giving numerical numbers as an encrypted information.

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fradulent transaction
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0->normal transaction

1->Fraudulent transaction

```python
#separating the data for analysis
```

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both of the classes
credit_card_data.drop('V13', axis=1).groupby('Class').mean()
```

Build a sample dataset containing similar distribution of normal transaction and fraudulent transactions

Number of fraudulent transactions->492

```python
legit_sample = legit.sample(n=492)
```

Concatenating the two dataframes

```python
import pandas as pd
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.drop('V13', axis=1).groupby('Class').mean()
```

Splitting the data into features & targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

Last step:Model Evaluation

Accuracy score

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data:',test_data_accuracy)
```
